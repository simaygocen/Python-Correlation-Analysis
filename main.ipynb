```python
# Importing the necessary packages and libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
pd.options.mode.chained_assignment = None
```

```python
# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Monster\Desktop\movies.csv')
```

```python
# Now let's take a look at the data
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
# We should remove the duplicate variables if there are any of them
df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Visualization of correlation between gross and budget to more clear understanding with seaborn
sns.regplot(x="gross", y="budget", data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# Visualization of correlation between score and gross to more clear understanding with seaborn
sns.regplot(x="score", y="gross", data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# Correlation Matrix between all numeric columns

numeric_columns = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_columns.corr(method='pearson')
print(correlation_matrix)
```

```python
correlation_matrix = numeric_columns.corr(method='kendall')
print(correlation_matrix)
```

```python
correlation_matrix = numeric_columns.corr(method='spearman')
print(correlation_matrix)
```

```python
correlation_matrix1 = correlation_matrix

sns.heatmap(correlation_matrix1, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
